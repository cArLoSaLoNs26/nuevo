```python
import numpy as np

# NumPy es una librería de Python especializada en el cálculo numérico y el análisis de datos, especialmente para un gran volumen de datos


#
## la siguiente función divide el test de datos en un conjunto de prueba y otro de entrenamiento
# @data: el set de datos
# @test_ratio: porciento del conjunto de instancias aleatorias seleccionadas (por lo general el 20% de los datos
# o menos si el conjunto es muy grande)
#

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data)) # realiza una nueva distribución aleatoria de la data. @len(data) es longitud de la data
    print(shuffled_indices)
    test_set_size = int(len(data) * test_ratio)# producto de la longitud de la data por la fracción de la data (generalmente 20%)
    test_indices = shuffled_indices[:test_set_size] # Los datos organizados aleatoriamente se dividen (slice 20%) desde 0 hasta el valor de test_set_size
    train_indices = shuffled_indices [test_set_size:] #Los datos organizados aleatoriamente se dividen (slice 80%) desde test_set_size hasta el valor final
    return data.iloc[train_indices], data.iloc[test_indices] #.iloc es un método de Pandas utilizado para la indexación basada en ubicaciones de números enteros. Le permite seleccionar filas y columnas por sus posiciones enteras
```
